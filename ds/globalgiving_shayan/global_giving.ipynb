```python
globalGiving.drop(globalGiving.columns[14:19], axis=1, inplace=True)
globalGiving.head()
```

Removed 4 data columns that included Nan Values.

```python
globalGiving.dtypes
```

```python
globalGiving.rename(columns={'Collected': 'Collected (in $)', 'Target': 'Target (in $)', 'Donations' : 'Donations (in $)'}, inplace=True)
```

```python
globalGiving.head()
```

-----Began converting number columns to float data types.-----

```python
globalGiving['Collected (in $)'] = globalGiving['Collected (in $)'].str.replace('$', '')
globalGiving['Target (in $)'] = globalGiving['Target (in $)'].str.replace('$', '')

globalGiving['Collected (in $)'].isna().sum()
globalGiving.head()
```

```python
globalGiving['Collected (in $)'].fillna(0, inplace=True)

globalGiving['Target (in $)'].isna().sum()
globalGiving['Target (in $)'].fillna(0,inplace=True)
```

```python
globalGiving['Donations (in $)'].isna().sum()
globalGiving['Donations (in $)'].fillna(0, inplace=True)

globalGiving['Target (in $)'] = globalGiving['Target (in $)'].str.replace(',', '')

```

```python
globalGiving['Target (in $)'].isna().sum()
```

```python
globalGiving['Target (in $)'].fillna(0, inplace=True)
```

```python
globalGiving['Target (in $)'] = pd.to_numeric(globalGiving['Target (in $)'], errors='coerce')
```

```python
globalGiving['Collected (in $)'] = globalGiving['Collected (in $)'].str.replace(',', '')
globalGiving['Collected (in $)'] = globalGiving['Collected (in $)'].astype('Int64')
```

```python
globalGiving['Donations (in $)'] = globalGiving['Donations (in $)'].astype('Int64')
```

-----Ended data cleaning for numerical data columns.-----

```python
globalGiving.dtypes
globalGiving.tail()
```

```python
globalGiving['Location'].value_counts(ascending=False)
```

```python
location_counts = globalGiving['Location'].value_counts()

# Slice the top 10 locations
top_10_locations = location_counts.head(10)

# Plot the bar chart using seaborn
sns.barplot(x=top_10_locations.index, y=top_10_locations.values, color='blue')

# Set labels and title
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.title('Top 10 locations with highest frequency of NGOs')
plt.xticks(rotation=90)
plt.show()
```
